# Augmentation

### for non flare lumnious images


```python
import os
import cv2
import numpy as np
from imgaug import augmenters as iaa

input_folder = '/content/drive/My Drive/Solar_Flare_Internship/non_flare_luminous'

# Creating augmentation pipeline
augmentation_pipeline = iaa.Sequential([
    iaa.Fliplr(0.5),  # Horizontal flips
    iaa.Flipud(0.5),  # Vertical flips
    iaa.Affine(rotate=(-30, 30)),  # Rotate images
    iaa.Affine(scale=(0.8, 1.2))  # Scale images
])

for filename in os.listdir(input_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):

        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        if image is None:
            print(f'Error loading image: {filename}')
            continue


        augmented_images = augmentation_pipeline(images=[image])


        for i, aug_image in enumerate(augmented_images):
            aug_image_path = os.path.join(input_folder, f'aug_{i}_{filename}')
            cv2.imwrite(aug_image_path, aug_image)

        print(f'Augmented and saved: {filename}')

print(f'Augmentation complete. Augmented images saved to the folder: {input_folder}')
```

# Labelling

```python
import os
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split


flare_folder = '/content/drive/My Drive/Solar_Flare_Internship/flare_luminous'
non_flare_folder = '/content/drive/My Drive/Solar_Flare_Internship/non_flare_luminous'

flare_images = [os.path.join(flare_folder, img) for img in os.listdir(flare_folder)]
non_flare_images = [os.path.join(non_flare_folder, img) for img in os.listdir(non_flare_folder)]

# Define labels (1 for flare, 0 for non-flare)
labels_flare = [1] * len(flare_images)
labels_non_flare = [0] * len(non_flare_images)

# Combine images and labels
images = flare_images + non_flare_images
labels = labels_flare + labels_non_flare
```

```python
# Shuffle the images and labels
combined = list(zip(images, labels))
np.random.shuffle(combined)
images, labels = zip(*combined)
```

# Normalization

```python
import os
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split

# Function to normalize an image
def normalize_image(image):
    normalized_image = np.array(image) / 255.0  # Normalize pixel values to [0, 1]
    return normalized_image


normalized_images = []


for img_path in images:
    try:
        img = Image.open(img_path)
        img_normalized = normalize_image(img)
        normalized_images.append(img_normalized)
    except Exception as e:
        print(f"Error processing {img_path}: {e}")
```

###Data generator method

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator


# Convert to NumPy arrays if not already
X_train = np.array(X_train)
y_train = np.array(y_train)

# Check shapes
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
X_test = np.array(X_test)
y_test = np.array(y_test)

X_val = np.array(X_val)
y_val = np.array(y_val)



print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
```

```python
image_size = (128, 128)
batch_size = 32

datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = datagen.flow(X_train, y_train, batch_size=batch_size)
validation_generator = datagen.flow(X_val, y_val, batch_size=batch_size)
test_generator = datagen.flow(X_test, y_test, batch_size=batch_size)
```

## CNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
```

```python

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', kernel_regularizer=l2(0.001), input_shape=(image_size[0], image_size[1], 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),


    Flatten(),
    Dense(512, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.5),

    Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.5),

    Dense(1, activation='sigmoid')  # Binary classification
])


model.compile(optimizer=Adam(learning_rate=0.00005), loss='binary_crossentropy', metrics=['accuracy'])

```

```python
# Train the model
history =model.fit(train_generator, epochs=150, validation_data=validation_generator)
```

```python
def load_random_image_from_folder(folder_path, image_size):

    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

    random_image_file = random.choice(image_files)


    image_path = os.path.join(folder_path, random_image_file)
    img = Image.open(image_path).resize(image_size)
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    return img_array, random_image_file
```
